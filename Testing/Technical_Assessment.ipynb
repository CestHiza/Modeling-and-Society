```python
#Drop the variables 'Urban', 'Unnamed: 0' and 'School Closure Date'
df1 = df.drop(columns = ['Urban', 'Unnamed: 0', 'School Closure Date'])
```

```python
#Sort all states by Population in order of greatest to least
df1.sort_values(by = 'Population', ascending = False)
```

```python
#Group states by region and calculate mean for the variables
#'Population', 'Tested', and 'Infected'
df2 = df1.groupby(by = 'Regions').sum()
df1.groupby(by = 'Regions')[['Population', 'Tested', 'Infected']].mean()
```

```python
#Create a new column that represents the Infected to Tested ratio that's called InfToTest
df1['InfToTest'] = df1['Infected']/df1['Tested']
df1['InfToTest']

#Oops! I want to rename the column to 'Inf : Test'
df1.rename(columns = {'InfToTest': 'Inf : Test'})
```
